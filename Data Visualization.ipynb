```python
#Data Visualization
#Line Plot
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 20, 12]
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```python
#Scatter Plot
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 20, 12]
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
#Barchar
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D']
values = [10, 30, 50, 25]
plt.bar(categories, values)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Plot')
plt.show()
```

```python
#Histogram
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot([0, 1, 2, 3, 4, 5])
plt.show()
```

```python
#Box Plot using Seaborn
import seaborn as sns
import numpy as np
data = np.random.normal(size=(100, 4))
sns.boxplot(data=data)
plt.title('Box Plot')
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C', 'D']
sizes = [20, 30, 25, 15]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([5,4,9,33])
plt.plot(x, marker = 'o')
plt.show()
```

```python
#Heat amp
import seaborn as sns
import numpy as np
data = np.random.rand(10, 12)
sns.heatmap(data)
plt.title('Heatmap')
plt.show()
```

```python
#Violin Plot
import seaborn as sns
import numpy as np
data = np.random.normal(size=(100, 5))
sns.violinplot(data=data)
plt.title('Violin Plot')
plt.show()
```
